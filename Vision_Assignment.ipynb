```python
import cv2 
import numpy as np
import statistics
from google.colab.patches import cv2_imshow
img1 = cv2.imread('Picture-1.PNG',1)
img2 = cv2.imread('Picture-2.PNG',1)
# Print error message if image is null
if img1 is None:
    print('Could not read image1')
if img2 is None:
    print('Could not read image2')
# Apply identity kernel or filter
kernel1 = np.array([[0, 0, 0],
                    [0, 1, 0],
                    [0, 0, 0]])
# Apply custom 2D kernel
kernel2 = np.ones((5, 5), np.float32) / 25
# Apply sharpening kernel
kernel3 = np.array([[0, -1,  0],
                   [-1,  5, -1],
                    [0, -1,  0]])
filtered_img1 = cv2.filter2D(src=img1, ddepth=-1, kernel=kernel1)
filtered_img2 = cv2.filter2D(src=img2, ddepth=-1, kernel=kernel1)
cv2_imshow(img1)
cv2_imshow(img2)
```

```python
cv2_imshow(filtered_img1)
cv2_imshow(filtered_img2)
```

```python
#Defining custom filter function to modify each pixcel value 
def filter_img (image):
  for i in range(image.shape[0]):
    for j in range(image.shape[1]):
      #if image[i][j][0] > 155:
      #  image[i][j][0] = 255
      #if image[i][j][1] > 155:
      #  image[i][j][1] = 255
      #if image[i][j][2] > 155:
      #  image[i][j][2] = 255 
      #modification of pixel values based on mean intensity values
      if statistics.mean(image[i][j]) > 150:
        image[i][j][:] = 255

  cv2_imshow(image)
```

```python
filter_img(img1)
filter_img(img2)
```
